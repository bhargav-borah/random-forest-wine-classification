## Data Preprocessing

```python
X = wine_dataset.drop(columns=['quality'])
y = wine_dataset['quality']
```

### Label Binarization

```python
y = y.apply(lambda y_val : 1 if y_val >= 7 else 0)
```

```python
y
```

```python
y.value_counts()
```

## Splitting the dataset into training set and test set

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, stratify=y, random_state=0)
```

## Training the Random Forest Classifier Model

```python
from sklearn.ensemble import RandomForestClassifier
classifier = RandomForestClassifier()
classifier.fit(X_train, y_train)
```

## Model Evaluation

```python
from sklearn.metrics import accuracy_score, confusion_matrix
```

### Accuracy score on training set

```python
y_pred_train = classifier.predict(X_train)
print('Accuracy score = {:.2f} %'.format(accuracy_score(y_train, y_pred_train)*100))
print(confusion_matrix(y_train, y_pred_train))
```

### Accuracy score on test set

```python
y_pred_test = classifier.predict(X_test)
print('Accuracy score = {:.2f} %'.format(accuracy_score(y_test, y_pred_test)*100))
print(confusion_matrix(y_test, y_pred_test))
```
